# Ejercicio Embrague



>Suponga que el sistema trabaja de manera horizontal. La fricción de las zapatas A contra el marco B hace funcional el embrague centrífugo y se transmite el torque de A hacia B. La fuerza centrífuga mantiene a las zapatas contra el marco B donde su coeficiente de fricción estático es de $\mu=0.8$. Cada zapata pesa 9.8[N] y el peso del resto de partes es despreciable. 
- Calcular la tensión inicial de los resortes (C) para que el contacto de las zapatas se haga cuando $\dot{\theta}\geq200[rpm]$
- Determinar el torque en el intervalo $100\geq\dot{\theta}\geq1500[rpm]$

$$\Sigma F \vec e_{r}=m a_{\vec e_{r}}$$

$$-N - F_e=m(\ddot{r}-r\dot{\theta}^2)$$

$$N+ F_e=mr\dot{\theta}^2$$

## Resolución
>Suponemos un eje de referencia inercial en el centro entre las dos zapatas, y a su vez se establece otro eje de referencia auxiliar en uno de los extremos de cualquier zapata, tal que se pueda realizar el análisis en base a dicho eje. 
El eje de refencia auxiliar, será un eje radial-transversal.
>

>Se procede analizar las fuerzas que actuan en cada eje, para poder realizar las sumatorias de fuerzas respectivas.
>Por lo tanto, en el eje radial, se tiene:

>

>Ahora, hay que entender que para que la zapata entre en conctacto con la cubierta, el valor de $N=0$, por lo tanto: 

$$F_e=mr\dot{\theta}^2$$

>Reemplazamos los valores del ejercicio:

$$F_e =(1kg)(0,3m)(200\frac{2\pi}{60})^2$$

$$F_e =131,59 [N]$$

>Ahora, para determinar el torque en el intervalo $100\geq\dot{\theta}\geq1500[rpm]$, se realiza el análisis de fuerzas en el eje transversal:

>

$$\Sigma F \vec e_{\theta}=m a_{\vec e_{\theta}}$$

$$-Fr=m(r\ddot{\theta}+2\dot{r}\dot{\theta})$$

$$-Fr=2m\dot{r}\dot{\theta}$$


>Ahora, haciendo uso de la ecuación calculada en el eje radial, podemos despejar el valor de $N$

$$N=mr\dot{\theta}^2 - F_e$$

>Usando los valores de $\dot{\theta}=1500[rpm]$ y el valor de la tensión inicial $Fe= 131,59[N]$, se obtiene:

$$N=(1)(0,3)(\frac{1500.2\pi}{60})^2 - 131,59$$

$$N=7270,61 [N]$$



>Con este valor, calculamos la fuerza de rozamiento:

$$Fr_e= u.N$$

$$Fr_e= (0,8)(7270,61)$$

$$Fr_e= 5816,49 [N]$$


>Podemos realizar el cálculo del torque a partir de este valor, en primer lugar hay que tomar en cuenta que para $\dot{\theta}= 100[rpm]$ el $\tau=0[Nm]$, ya que en este punto la carcaza aún no entra en contacto con las zapatas.
>Ahora, para $\dot{\theta}= 1500[rpm]$, usamos la $Fr_e$ encontrada previamente:

$$\tau= Fr_e.d$$

$$\tau= (5816,49).(0,4)$$

$$\tau=2326,596 [Nm]$$

>Pero hay que tomar en cuenta que existen dos zapatas que realizan torque, por lo que el torque en $\dot{\theta}= 1500[rpm]$, es:

$$\tau=2326,596 . 2$$

$$\tau=4653,19 [Nm]$$
